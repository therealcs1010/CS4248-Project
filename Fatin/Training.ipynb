```python
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import roc_curve, auc
```

```python
# Get predicitons
def get_preds(test_feat, train_feat, y_test, y_train, model, title, file):
    model.fit(train_feat, y_train)
    preds = model.predict(test_feat)
    fpr, tpr, _ = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)
    print('AUC:', roc_auc)
    plt.plot(fpr, tpr)
    plt.title(title)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
    plt.savefig('data/' + file + '.png')
    return preds
```

```python
from numpy import loadtxt
import pandas as pd
```

```python
path = 'pre-processed_data/'
train_bow = loadtxt(path + 'train_bow.csv', delimiter=',')
test_bow = loadtxt(path + 'test_bow.csv', delimiter=',')
train_sentiment = pd.read_csv(path + 'train_sentiment.csv', sep=',')  
test_sentiment = pd.read_csv(path + 'test_sentiment.csv', sep=',')  
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, ComplementNB, BernoulliNB, CategoricalNB
```

```python
preds_rf = get_preds(test_bow, train_bow, 
                  test_sentiment['label'], train_sentiment['label'], 
                  DecisionTreeClassifier(random_state=4248), 
                  'Decision Tree (with BOW)', 'decision_tree_bow')
```

```python
preds_rf = get_preds(test_bow, train_bow, 
                  test_sentiment['label'], train_sentiment['label'], 
                  RandomForestClassifier(random_state=4248),
                  'Random Forest (with BOW)', 'random_forest_bow')
```
