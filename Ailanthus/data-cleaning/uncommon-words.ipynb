```python
vocab = Counter()

def add_to_vocab(text):
  text = text.split(" ")
  a = []
  for word in text:
    if len(word) <= 2:
      continue
    else:
      a.append(word)

  vocab.update(a)

train_df["text"].apply(lambda x: add_to_vocab(x))
```

```python
print(len(vocab))
print(vocab.most_common(50))
n = len(vocab)

# least common
print(vocab.most_common()[:-100:-1])
```

```python
min = 3
cleaned_vocab = [key for key, value in vocab.items() if value >= min]

print(len(cleaned_vocab))
```

```python
def remove_uncommon_words(text):
  text_arr = text.split()
  x = []
  for word in text_arr:
    if word in cleaned_vocab:
      x.append(word)
  
  text = " ".join(x)

  return text
```

```python
train_df["text"] = train_df["text"].apply(lambda x: remove_uncommon_words(x))
```

```python
test_df["text"] = test_df["text"].apply(lambda x: remove_uncommon_words(x))
```
